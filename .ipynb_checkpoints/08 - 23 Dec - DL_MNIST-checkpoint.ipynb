**1. Importing reuquired libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization, Flatten
```

**2. Loading the MNIST Dataset**

```python
from keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test) = mnist.load_data()
```

```python
x_train.shape
```

```python
x_test.shape
```

**Obseervations**

*   Train data has 60000 images of 28X28 dimensions
*   Test data has 10000 images of 28 X 28 dimesions
*   The images are in black and white format

```python
# Viewing few random images from the data
plt.figure(figsize=(10,10))
np.random.seed(0)
index = np.random.randint(0,60000,9)
for i in range(len(index)):
  plt.subplot(3,3,i+1)
  plt.imshow(x_train[index[i]],cmap='gray')
```

**3. Data Preprocessing**

```python
# Min - max scaling without using libarary
# x_min = 0 (The smallest pixel value)
# x_max = 255(The largest pixel value)

# Formula min_max scaling : (x-x_min)/(x_max-x_min)
#=> x/x_max

x_train_scaled = x_train/255
x_test_scaled = x_test/255
```

```python
x_train_flattened = x_train_scaled.reshape(60000,784)
x_test_flattened = x_test_scaled.reshape(10000,784)
```

```python
print(y_train[0])
```

```python
# Converting the output to One hot Encoding using to_categorical
y_train_encoded = to_categorical(y_train,num_classes=10)
y_test_encoded = to_categorical(y_test,num_classes=10)
```

```python
print(y_train_encoded[0])
```

# **4. Define the model**

```python
from tensorflow.keras.layers import Input
model = Sequential()
model.add(Input(shape = (784,)))
model.add(Dense(10,activation="softmax"))
```

```python
model.summary()
```

**5. Compile the model**

```python
model.compile(optimizer='sgd',loss="categorical_crossentropy",metrics = ['accuracy'])
```

# **6. Fit the Model**

```python
model.fit(x_train_flattened,y_train_encoded, batch_size = 128, epochs = 10)
```

# **7. Evaluate the Model**

```python
model.evaluate(x_test_flattened,y_test_encoded)
```

```python
model.evaluate(x_test_flattened,y_test_encoded,verbose = 0)
```
